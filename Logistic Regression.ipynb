## Aim:  What is Logistic Regression?

#### When used for classification, Logistic Regression gives an output of 0/1 or Yes/No, Do/Die etc...

#### But we can use it for classification

Logistic Regression gives a probablity for particular output. If the probability is less than certain threshold(which we can set) we can ouput as 'No' or '0'..otherwise we can output 'Yes' or '1'.

So consider that for a particular configuration of universe, we got the output probaility as 0.75. That means that there are high chances that the universe will sustain (because it's more than 50% so more likely to sustain), so we can output as 'Yes' or '1'.

#### Why is it called as 'Regression' then?

Because it gives a continuous output i.e probability between 0 and 1
